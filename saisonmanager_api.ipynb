```python
import requests
import pandas as pd

url = "https://saisonmanager.de/api/v2/leagues/1890/scorer.json"
response = requests.get(url)
response.raise_for_status()

data = response.json()

# Spieler des Teams filtern
team_name = "SSF Dragons Bonn"
dragons = [p for p in data if p.get("team_name") == team_name]

# Absteigend nach penalty_2 sortieren
dragons_sorted = sorted(dragons, key=lambda p: p.get("penalty_2", 0), reverse=True)

df = pd.DataFrame(dragons_sorted)

# Spalten filtern: first_name, last_name und alle die mit "penalty" beginnen
penalty_cols = [c for c in df.columns if c.startswith("penalty")]
df[["first_name", "last_name"] + penalty_cols].head(10)
```
